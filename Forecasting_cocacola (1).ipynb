```python
quarter=['Q1','Q2','Q3','Q4']
n=data['Quarter'][0]
n[0:2]

data['quarter']=0
```

```python
for i in range(42):
    n=data['Quarter'][i]
    data['quarter'][i]=n[0:2]
```

```python
dummy=pd.DataFrame(pd.get_dummies(data['quarter']))
dummy.head()
```

```python
coco=pd.concat((data,dummy),axis=1)
t= np.arange(1,43)
coco['t']=t
coco['t_square']=coco['t']*coco['t']
coco
```

```python
log_Sales=np.log(coco['Sales'])
coco['log_Sales']=log_Sales
coco.head()
```

```python
Train= coco.head(38)
Test=coco.tail(4)
```

##### Linear model

```python
import statsmodels.formula.api as smf 

linear_model = smf.ols('Sales~t',data=Train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(Test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_linear))**2)) 
rmse_linear
```

#### Exponential model

```python
Exp_model = smf.ols('log_Sales~t',data=Train).fit()
pred_Exp=  pd.Series(Exp_model.predict(pd.DataFrame(Test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Exp)))**2)) 
rmse_Exp
```

#### Quadratic model

```python
Quad_model= smf.ols('Sales~t+t_square',data=Train).fit()
pred_Quad = pd.Series(Quad_model.predict(Test[["t","t_square"]]))
rmse_Quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad 
```

#### Additive seasonality 

```python
add_sea = smf.ols('Sales~ Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea = pd.Series(add_sea.predict(Test[['Q1','Q2','Q3','Q4']]))
rmse_add_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

#### Additive seasonality with Quadratic

```python
add_sea_Quad = smf.ols('Sales~t+t_square+Q1+Q2+Q3+Q4',data=Train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(Test[['t','t_square','Q1','Q2','Q3','Q4']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

#### Multiplicative seasonality

```python
Mul_sea = smf.ols('log_Sales~Q1+Q2+Q3+Q4',data = Train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(pd.DataFrame(Test[['Q1','Q2','Q3','Q4']])))
rmse_Mult_sea = np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

#### Multiplicative seasonality with quadratic 

```python
mul_quad= smf.ols('log_Sales~t+t_square+Q1+Q2+Q3+Q4',data=Train).fit()
pred_mul_quad= pd.Series(mul_quad.predict(Test[['t','t_square','Q1','Q2','Q3','Q4']]))
rmse_mul_quad=np.sqrt(np.mean((np.array(Test['Sales'])-np.array(np.exp(pred_mul_quad)))**2))
rmse_mul_quad
```

```python
Models = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_mul_quad"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_mul_quad])}
table_rmse=pd.DataFrame(Models)
table_rmse.sort_values(['RMSE_Values'])
```

#### From above rmse values we can consider Additional seasonality with quadratic trend model
   The data seems to be volatile i.e, it is unstable .
   We can say volatile data can make it difficult to accurately forecast future values, as the data may be subject to sudden, unexpected changes that cannot be easily predicted.

### Evaluation metric MAPE

```python
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import SimpleExpSmoothing # SES
from statsmodels.tsa.holtwinters import Holt # Holts Exponential Smoothing
from statsmodels.tsa.holtwinters import ExponentialSmoothing  
```

### Simple exponential method

```python
ses_model = SimpleExpSmoothing(Train['Sales']).fit(smoothing_level=0.2)
pred_ses = ses_model.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_ses,Test.Sales) 
```

### Holts method

```python
hw_model = Holt(Train['Sales']).fit(smoothing_level=0.8, smoothing_slope=0.2)
pred_hw = hw_model.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hw,Test.Sales) 
```

### Holts winter exponential smoothing with additive seasonality and additive trend

```python
hwe_model_add_add = ExponentialSmoothing(Train["Sales"],seasonal="add",trend="add",seasonal_periods=12).fit() #add the trend to the model
pred_hwe_add_add = hwe_model_add_add.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hwe_add_add,Test.Sales) 
```

### Holts winter exponential smoothing multiplicative seasonality and additive trend

```python
hwe_model_mul_add = ExponentialSmoothing(Train["Sales"],seasonal="mul",trend="add",seasonal_periods=12).fit() 
pred_hwe_mul_add = hwe_model_mul_add.predict(start = Test.index[0],end = Test.index[-1])
MAPE(pred_hwe_mul_add,Test.Sales)
```

##### From above we consider Holts winter exponential smoothing with multiplicative seasonality and additive trend since its MAPE value is low.

### Final model by combining train and test data

```python
hwe_model_mul_add_final = ExponentialSmoothing(coco["Sales"],seasonal="mul",trend="add",seasonal_periods=12).fit() 
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
  
model = SARIMAX(Train['Sales'], 
                order = (0, 1, 0), 
                seasonal_order =(0,1,0,4))
  
result = model.fit()
result.summary()
```

```python
start = len(Train)
end = len(Train) + len(Test) - 1
  
# Predictions for one-year against the test set
predictions = result.predict(start, end,
                             typ = 'levels').rename("Predictions")
  
# plot predictions and actual values
predictions.plot(legend = True)
Test['Sales'].plot(legend = True)
```

```python
from sklearn.metrics import mean_squared_error
from statsmodels.tools.eval_measures import rmse
  
rmse(Test["Sales"], predictions)
  
mean_squared_error(Test["Sales"], predictions)
```
